# Automatyczne wyszukiwanie parametrów

https://scikit-learn.org/stable/getting_started.html

https://scikit-learn.org/stable/modules/grid_search.html#grid-search

https://scikit-learn.org/stable/modules/model_evaluation.html

Wszystkie estymatory posiadają parametry (często nazywane **hiperparametrami** w literaturze), które można dostosować. 

Ogólna siła uogólniania estymatora często zależy krytycznie od kilku parametrów. Na przykład **RandomForestClassifier** posiada parametr **n_estimators**, który określa liczbę drzew w lesie, oraz parametr **max_depth**, który określa maksymalną głębokość każdego drzewa. Często nie jest jasne, jakie dokładnie powinny być wartości tych parametrów, ponieważ zależą one od dostępnych danych.

**Scikit-learn** dostarcza narzędzi do automatycznego znalezienia najlepszych kombinacji parametrów (poprzez walidację krzyżową). W poniższym przykładzie przeszukujemy losowo przestrzeń parametrów lasu losowego za pomocą obiektu **RandomizedSearchCV**. Po zakończeniu przeszukiwania, **RandomizedSearchCV** zachowuje się jak **RandomForestClassifier**, który został dopasowany do najlepszego zestawu parametrów.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from scipy.stats import randint

from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv('.\\Data\\kyphosis.csv')
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size = 0.3)

# Zdefiniuj przestrzeń parametrów, która będzie poddana przeszukiwaniu.
param_distributions = {'n_estimators': randint(10, 200),
                       'max_depth':    randint(5, 10)}

# Teraz stwórz obiekt SearchCV i dopasuj go do danych.
search = RandomizedSearchCV(estimator=RandomForestClassifier(random_state=0),
                            n_iter=25,
                            param_distributions=param_distributions,
                            random_state=0,
                            scoring = 'accuracy')
search.fit(X_train, y_train)
search.best_params_
```

```python
search.score(X_test, y_test)
```

Przykład z dataset 'iris' i zmianą parametrów modelu (GridSearchCV):

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV
iris = datasets.load_iris()
parameters = {'kernel':('linear', 'rbf'), 'C':[1, 10]}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(iris.data, iris.target)
```

```python
clf.best_params_
```

```python
sorted(clf.cv_results_.keys())
```

```python
clf.best_score_
```

# Granice decyzyjne

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from IPython.display import Image
from sklearn.tree import DecisionTreeClassifier
sns.set()
```

```python
iris = load_iris()
data = iris.data
targets = iris.target
feature_names = [name.replace(' ', '_')[:-5] for name in iris.feature_names]
class_names = iris.target_names

print('Liczba próbek:', len(data))
print('Kształt danych', data.shape)
print('Nazwy zmiennych objaśniających:', feature_names)
print('Nazwy kategorii kosaćca:', class_names)
```

```python
print(iris.DESCR)
```

```python
data_targets = np.c_[data, targets]
df = pd.DataFrame(data_targets, columns=feature_names + ['class'])
df.head()
```

```python
df.describe().apply(lambda x: round(x, 2))
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='class', data=df, legend='full')
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x = 'petal_length', y = 'petal_width', hue='class', data=df, legend='full')
```

```python
from sklearn.tree import export_graphviz
from mlxtend.plotting import plot_decision_regions
from six import StringIO
from IPython.display import Image
import pydotplus
```

```python
def make_decision_tree(max_depth=1):
    # train model
    tree_clf = DecisionTreeClassifier(max_depth=max_depth, random_state=30)
    tree_clf.fit(X, y)

    # export graph
    dot_data = StringIO()
    export_graphviz(tree_clf,
                   out_file=dot_data,
                   feature_names=feature_names[2:],
                   class_names=class_names,
                   special_characters=True,
                   rounded=True,
                   filled=True)
    graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
    graph.write_png('graph.png')
    
    
    # plot decision boundries
    colors='#f1865b,#31c30f,#64647F,#d62728,#9467bd,#8c564b,#e377c2,#7f7f7f,#bcbd22,#17becf'
    plt.figure(figsize=(8, 6))
    ax = plot_decision_regions(X, y, tree_clf, legend=0, colors=colors)
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, ['setosa', 'versicolor', 'virginica'], framealpha=0.3)
    plt.xlabel('petal length (cm)')
    plt.ylabel('petal width (cm)')
    plt.title('Drzewo decyzyjne: max_depth={}'.format(max_depth))
    
    # accuracy
    print('Dokładność modelu: ' + str(round(tree_clf.score(X, y) * 100, 2)) + '%')
    return Image(graph.create_png())
```

## Przygotowanie danych do modelu

```python
X = df.copy()
X = X[['petal_length', 'petal_width', 'class']]
y = X.pop('class')
```

```python
# Konwersja DataFrame do tablic numpy
X = X.values
y = y.values.astype('int16')
```

## Tworzenie modeli

```python
make_decision_tree()
```
