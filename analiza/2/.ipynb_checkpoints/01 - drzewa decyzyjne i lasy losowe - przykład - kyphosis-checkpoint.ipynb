# Drzewa decyzyjne i lasy losowe

**Drzewa decyzyjne** (ang. decision trees), zwane równiez drzewami klasyfikacyjnymi (ang. classification trees), opierają się o drzewiastą strukturę.

**Lasy losowe** (random forests) będące uogólnieniem idei drzew decyzyjnych zalicza się do procedur agregujących (ensemble method). Działanie lasów losowych polega na klasyfikacji za pomocą grupy drzew decyzyjnych. Koncowa decyzja jest podejmowana w wyniku głosowania większościowego nad klasami wskazanymi przez poszczególne drzewa decyzyjne.

## Dane

Dane zebrano na 83 pacjentach poddanych korekcji chirurgii kręgosłupa. Celem było określenie ważnych czynników ryzyka
na kifozę po operacji. Czynnikami ryzyka są wiek w miesiącach, pierwszy z kręgów i liczby zaangażowanych poziomów.
Przypadki 15 i 28 są usuwane, w związku z tym zostaje 81 przypadków.

Kolumny:
* Age = miesiące
* Start = początek kręgów
* Number = liczba kręgów
* Kyphosis = kifoza, present jeśli jest obecna, absent jeśli nieobecna

Kifoza pogłębiona (inaczej plecy okrągłe) to choroba kręgosłupa, której cechą charakterystyczną jest nadmierne wygięcie kręgosłupa ku tyłowi w odcinku piersiowym i krzyżowym.

## Zestaw treningowy i zestaw testowy

Podzielmy dane na zestaw treningowy i zestaw testowy!

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.30)
```

## Drzewa decyzyjne

Zaczniemy od treningu pojedynczego drzewa decyzyjnego.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Predykcja

```python
# bez index'u będzie błąd "If using all scalar values, you must pass an index"
data = pd.DataFrame(data={'Age': 34, 'Number': 4, 'Start':3},index=[0]) 
data
```

```python
# lub
data = pd.DataFrame(data={'Age':[34], 'Number': [4], 'Start':[3]})
data
```

```python
predictions = dtree.predict(data)
print('Kyphosis = ' + predictions[0])
```

## Prognozowanie i ocena

Przeanalizujmy nasze drzewo decyzyjne.

```python
 dtree_pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test, dtree_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,  dtree_pred)
```

```python
print(confusion_matrix(y_test, dtree_pred))
```

```python
cm = confusion_matrix(y_test, dtree_pred)
sns.heatmap(cm, annot=True)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
feature_importances = pd.DataFrame(dtree.feature_importances_,
                                   index = X_train.columns,
                                   columns=['importance']).sort_values('importance',ascending=False)
feature_importances
```

# Dodatek (Automatyczne wyszukiwanie parametrów)

https://scikit-learn.org/stable/getting_started.html

https://scikit-learn.org/stable/modules/grid_search.html#grid-search

https://scikit-learn.org/stable/modules/model_evaluation.html

Wszystkie estymatory posiadają parametry (często nazywane **hiperparametrami** w literaturze), które można dostosować. 

Ogólna siła uogólniania estymatora często zależy krytycznie od kilku parametrów. Na przykład **RandomForestClassifier** posiada parametr **n_estimators**, który określa liczbę drzew w lesie, oraz parametr **max_depth**, który określa maksymalną głębokość każdego drzewa. Często nie jest jasne, jakie dokładnie powinny być wartości tych parametrów, ponieważ zależą one od dostępnych danych.

**Scikit-learn** dostarcza narzędzi do automatycznego znalezienia najlepszych kombinacji parametrów (poprzez walidację krzyżową). W poniższym przykładzie przeszukujemy losowo przestrzeń parametrów lasu losowego za pomocą obiektu **RandomizedSearchCV**. Po zakończeniu przeszukiwania, **RandomizedSearchCV** zachowuje się jak **RandomForestClassifier**, który został dopasowany do najlepszego zestawu parametrów.

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import train_test_split
from scipy.stats import randint

X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size = 0.3)

# Zdefiniuj przestrzeń parametrów, która będzie poddana przeszukiwaniu.
param_distributions = {'n_estimators': randint(10, 200),
                       'max_depth':    randint(5, 10)}

# Teraz stwórz obiekt SearchCV i dopasuj go do danych.
search = RandomizedSearchCV(estimator=RandomForestClassifier(random_state=0),
                            n_iter=25,
                            param_distributions=param_distributions,
                            random_state=0,
                            scoring = 'accuracy')
search.fit(X_train, y_train)
search.best_params_
```

```python
search.score(X_test, y_test)
```
