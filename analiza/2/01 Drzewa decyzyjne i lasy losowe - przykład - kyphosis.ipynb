# Drzewa decyzyjne i lasy losowe

**Drzewa decyzyjne** (ang. decision trees), zwane równiez drzewami klasyfikacyjnymi (ang. classification trees), opierają się o drzewiastą strukturę.

**Lasy losowe** (random forests) będące uogólnieniem idei drzew decyzyjnych zalicza się do procedur agregujących (ensemble method). Działanie lasów losowych polega na klasyfikacji za pomocą grupy drzew decyzyjnych. Koncowa decyzja jest podejmowana w wyniku głosowania większościowego nad klasami wskazanymi przez poszczególne drzewa decyzyjne.

## Dane

Dane zebrano na 83 pacjentach (dzieciach) poddanych korekcji chirurgii kręgosłupa. Celem było określenie ważnych czynników ryzyka na kifozę po operacji. Czynnikami ryzyka są wiek w miesiącach, pierwszy z kręgów i liczby zaangażowanych poziomów.
Przypadki 15 i 28 zostały usunięta, w związku z tym zostaje 81 przypadków.

Kolumny:
* Age = miesiące
* Start = numer pierwszego (najwyżej) operowanego kręgu.
* Number = liczba zaangażowanych kręgów
* Kyphosis = kifoza, present jeśli jest obecna, absent jeśli nieobecna

Kifoza pogłębiona (inaczej plecy okrągłe) to choroba kręgosłupa, której cechą charakterystyczną jest nadmierne wygięcie kręgosłupa ku tyłowi w odcinku piersiowym i krzyżowym.

## Zestaw treningowy i zestaw testowy

Podzielmy dane na zestaw treningowy i zestaw testowy!

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Kyphosis',axis=1)
y = df['Kyphosis']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.30)
```

```python
X_test.describe()
```

```python
X_train.describe()
```

## Drzewa decyzyjne

Zaczniemy od treningu pojedynczego drzewa decyzyjnego.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

## Predykcja

```python
# bez index'u będzie błąd "If using all scalar values, you must pass an index"
data = pd.DataFrame(data={'Age': 34, 'Number': 4, 'Start':3},index=[0]) 
data
```

```python
# lub
data = pd.DataFrame(data={'Age':[34], 'Number': [4], 'Start':[3]})
data
```

```python
predictions = dtree.predict(data)
print('Kyphosis = ' + predictions[0])
```

## Prognozowanie i ocena

Przeanalizujmy nasze drzewo decyzyjne.

```python
 dtree_pred = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test, dtree_pred))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,  dtree_pred)
```

```python
print(confusion_matrix(y_test, dtree_pred))
```

```python
cm = confusion_matrix(y_test, dtree_pred)
sns.heatmap(cm, annot=True)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
feature_importances = pd.DataFrame(dtree.feature_importances_,
                                   index = X_train.columns,
                                   columns=['importance']).sort_values('importance',ascending=False)
feature_importances
```

## Wizualizacja drzewa (wersja 2)

```python
import pydotplus
```

```python
df["Kyphosis"] = df["Kyphosis"].astype("category")
class_names = df.Kyphosis.cat.categories
feature_names = list(df.columns[1:])
```

```python
# export graph
dot_data = StringIO()
export_graphviz(dtree,
                out_file=dot_data,
                feature_names=feature_names,
                class_names=class_names,
                special_characters=True,
                rounded=True,
                filled=True)
```

```python
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('graph.png')
Image(graph.create_png())
```

## Losowe lasy

Teraz porównajmy model drzewa decyzyjnego z losowym lasem.

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100)
rfc.fit(X_train, y_train);
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
accuracy_score(y_test, rfc_pred)
```

```python
print(confusion_matrix(y_test,rfc_pred))
```

```python
cm = confusion_matrix(y_test, rfc_pred)
sns.heatmap(cm, annot=True)
plt.xlabel('true label')
plt.ylabel('predicted label');
```

```python
print(classification_report(y_test,rfc_pred))
```
